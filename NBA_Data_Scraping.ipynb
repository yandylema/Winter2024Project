```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://basketball.realgm.com/ncaa/stats/2024/Averages/All/All/Season/All/points/desc/1/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find('table')
# <table class="tablesaw compact tablesaw-swipe" data-tablesaw-mode="swipe" data-tablesaw-mode-switch="" data-tablesaw-mode-exclude="columntoggle" id="table-1564" style="">
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  # print(individual_row_data)
  length = len(df)
  df.loc[length] =individual_row_data
```
